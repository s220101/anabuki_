```python
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
x_ax = 1
y_ax = 1

color="#FF0000"
title = "タイトル"
xlabel = "横軸"
ylabel = "縦軸"
font_size = 14
font_family = 'MS Gothic'
figsize = (10,6)
line_width = 2
rotation = 0

plt.title(title)
plt.xlabel(xlabel)
plt.ylabel(ylabel)

plt.xlim()
plt.ylim()

plt.rcParams["font.size"] = font_size
plt.rcParams["font.family"] = font_family
plt.rcParams["figure.figsize"] == figsize

plt.xticks(rotation=rotation)

plt.plot(x_ax,y_ax,color=color,label=title,linewidth=line_width)
plt.legend(ncol=1)
plt.grid()
plt.show()
```
