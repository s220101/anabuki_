```python
import matplotlib.pyplot as plt
import japanize_matplotlib
import pandas as pd
import os


#変数の間
#csvファイルの番号
csv_columns_number = 0
#カラムの番号
columns_number = 2


#関数の間
#同一フォルダ内のcsvファイルの名前を取得
def get_csv_files_in_same_folder():

    # 現在のノートブックのディレクトリを取得 
    notebook_directory = os.path.dirname(os.path.abspath(r"ここにcsvファイルの絶対パスを記入"))
    # フォルダ内のファイル一覧を取得
    files = os.listdir(notebook_directory)

    # CSVファイルのみを抽出
    csv_files = [file for file in files if file.endswith(".csv")]

    return csv_files

def read_csv(csv_files_list):

    # ファイル名を保存する配列を作成（最低限の大きさで初期化）
    file_names_array = [None] * len(csv_files_list)

    # 配列にファイル名を格納
    for i, csv_file in enumerate(csv_files_list):
        file_names_array[i] = csv_file
    
    #csvの読込み
    read_df = [0]*len(file_names_array)
    for i in range(4):
        read_df[i] = pd.read_csv(file_names_array[i])
    return read_df

def csv_ploter(csv_df,record=2,title='Default',save_flg=False):
    
    #変数の間
    color="#0000FF"
    title = title
 
    xlabel = csv_df.columns[0]
    ylabel = csv_df.columns[record]
    font_size = 14
    font_family = 'MS Gothic'
    figsize = (18,6)
    line_width = 2
    rotation = 0
    
    #x軸は固定
    x_axis = csv_df[csv_df.columns[0]]
    #変化するのはY軸
    y_axis = csv_df[csv_df.columns[record]]

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    plt.xlim()
    plt.ylim()

    plt.rcParams["font.size"] = font_size
    plt.rcParams["font.family"] = font_family
    plt.rcParams["figure.figsize"] = figsize

    plt.xticks(rotation=rotation)

    plt.plot(x_axis,y_axis,color=color,label=ylabel,linewidth=line_width)
    plt.legend(ncol=1)
    plt.grid()
    #プロットした画像を保存する
    if save_flg == True:
        plt.savefig(str(csv_df.columns) + ".jpg")
    plt.show()


#メイン処理
# CSVファイルの一覧を取得
csv_files_list = get_csv_files_in_same_folder()

#csvファイルのpandasを利用した読込み
df = read_csv(csv_files_list)

#対象のデータフレームのカラム数よりcolumns_numberが少ない場合：True
while len(df) >= csv_columns_number+1:

    if columns_number == 2:
        #データのカラムを出力
        print(df[csv_columns_number].columns)
        print("カラム数:" + str(csv_columns_number))
    
    #対象のデータフレームのカラム名を出力
    print("calumns_name:" + str(df[csv_columns_number].columns[columns_number]))

    #csvファイルのデータをプロット
    csv_ploter(df[csv_columns_number][df[csv_columns_number].columns],record=columns_number,title=df[csv_columns_number].columns[columns_number],save_flg=False)

    columns_number += 1

    #データフレームの最後のカラムまでいったら
    if len(df[csv_columns_number].columns) <= columns_number:


        #次のcsvファイルを対象にし、その中の対象カラム番号を2にする
        csv_columns_number += 1
        columns_number = 2

    else:
        print()

print("終了")
```
