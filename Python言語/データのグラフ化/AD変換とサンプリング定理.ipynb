```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import japanize_matplotlib
```

```python
A = 1    # 振幅
f = 1.0    # 周波数 Hz
sec = 1.0  # 信号の長さ s
sf = 60# サンプリング周波数 Hz
plt.rcParams["font.size"] = 25
t = np.arange(0, sec, 1/sf) #サンプリング点の生成
plt.title("アナログ信号")
plt.xlabel("時間(t)")
plt.ylabel("振幅")
plt.grid()
y = np.sin(2*np.pi*f*t) # 正弦波の生成
plt.rcParams["figure.figsize"] = (12, 8)
#plt.scatter(t,y,color='#FF0000')
plt.plot(t,y,color='#3232FF')
fig = plt.figure()
plt.show()
#plt.plot(t, y)
```

```python
A = 1.0    # 振幅
f = 1    # 周波数 Hz
sec = 1.01  # 信号の長さ s
sf = 40 # サンプリング周波数 Hz

t = np.arange(0, sec, 1/sf) #サンプリング点の生成
plt.rcParams["font.size"] = 10
plt.rcParams["figure.figsize"] = (6, 4)
plt.title("アナログ信号のデジタル化")
plt.xlabel("時間(s)")
plt.ylabel("振幅")
plt.grid()
y = A * np.sin(2*np.pi*f*t) # 正弦波の生成
plt.plot(t,y,color='#3232FF',label='アナログ')
plt.scatter(t,y,color='#FF0000',label='デジタル')
plt.legend(ncol=1)
plt.show()
```

```python
A = 1.0    # 振幅
f = 1.0   # 周波数 Hz
sec = 1.001  # 信号の長さ s
sf = 36000 # サンプリング周波数 Hz
Δt = 1.9 # サンプリング周波数 Hz
sf_d = Δt# * 2# サンプリング周波数 Hz
plt.rcParams["font.size"] = 10
plt.rcParams["figure.figsize"] = (6, 4)
t = np.arange(0, sec, 1/sf) #サンプリング点の生成
t_d = np.arange(0, sec, 1/sf_d) #サンプリング点の生成
plt.title("周波数:" + str(f) + "Hz," + "サンプリング周波数:" + str(Δt) + "Hz")
plt.xlabel("時間(s)")
plt.ylabel("振幅")
plt.xlim(-0.1,1.1)
plt.ylim(-1.3,1.4)
plt.grid()

y = A * np.sin(2*np.pi*f*t) # 正弦波の生成
y_d = A * np.sin(2*np.pi*f*t_d) *1# 正弦波の生成
plt.plot(t,y,color='#3232FF',label='アナログ')
plt.plot(t_d,y_d,color='#FF0000',label='デジタル')
#plt.scatter(t_d,y_d,color='#FF0000',label='デジタル')
plt.legend(ncol=3)
plt.show()
```
